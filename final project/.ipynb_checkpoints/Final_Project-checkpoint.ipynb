```python
def corr_plot(data, intervals, annot=False):
    '''
        data - Pandas DataFrame with numerical data for correcaltion 
        calc and plot it in seaborn.heatmap
        
        interval - list of turples [(v_min_1, v_max_1)......(v_min_n, v_max_n)]
    '''
    
    
    plots = len(intervals)
    
    if plots == 1:
        plt.figure(figsize=(16, 16))
        seaborn.heatmap(data.corr().abs(), square=True, vmin=intervals[0][0], 
                        vmax=intervals[0][1], annot=annot) 
        
    else:
        fig, axes = plt.subplots(ncols=1, nrows=plots)
        fig.set_figheight(plots * 16)
        fig.set_figwidth(16)

        for (vmin, vmax), ax in zip(intervals, axes):
            seaborn.heatmap(data.corr().abs(), square=True, vmin=vmin, 
                            vmax=vmax, ax=ax, annot=annot)
```

# Model selection and fitting

## Several function for convenient further work 

```python
def classifier_name(estimator):
    '''
        crutch for getting name of sklearn.estimator class
    '''
    return str(estimator).split('.')[-1][:-2]
```

```python
def features_get(estimators, x, y, **kwargs):
    
    '''
        estimators - estimator class obj for selecting features
        x, y - features, target value
        params - parameters for estimators
        kwargs - paramaters for function controlling and estimators
    '''
    
    report = dict()
    best_estimators = dict()
    
    n_features = 10
    rfe_step = 3
    
    if 'n_features' in kwargs:
        n_features = kwargs['n_features']
    
    if 'rfe_step' in kwargs:
        rfe_step = kwargs['rfe_step']
    
    for estimator in tqdm.tqdm(estimators):
        
        estimator_name = classifier_name(estimator=estimator)
        
        param = dict()
        if estimator_name + '_params' in kwargs:
            param = kwargs[estimator_name + '_params']
        
        
        regressor = estimator(**param)
        
        rfe = RFE(regressor, step=rfe_step, n_features_to_select=n_features)
        rfe.fit(x, y)
        report[estimator_name + '_features'] = list(train.columns[np.nonzero(rfe.support_)])
        best_estimators[estimator_name] = rfe.estimator_    
    
    report = pd.DataFrame(data=report)
    
    return {
        'report': report,
        'estimators': best_estimators
    }       
```

```python
def split_data_to_classes(data, column='', new_column='', values=[]):
    '''
        data - pandas DataFrame
        column - name of a column contraining data compared with value from values
        new_column - contains comparison results
        values - contains values for comparison
    '''
    data = deepcopy(data)
    
    def func(x):
        for value, cluster in zip(reversed(values), range(1, len(values))):
            if x > value:
                return cluster
                                    
        return 0
            
    data[new_column] = data[column].apply(func)
                                    
    return data
```

```python
def regressor_prepaire(estimator, grid_params, X, Y):
    '''
        estimator - sklearn class implementing fit and predict methods
        grid_params - set of various parameters for GridSearch
        estimator_name - readable name for pandas DataFrame
        X, Y - data for fitting and validating
    '''
    
    estimator_name = classifier_name(estimator)
    
    grid_search = GridSearchCV(estimator(), grid_params, n_jobs=4, iid=True, cv=25)
    grid_search.fit(X, Y)
    
    best_estimator = grid_search.best_estimator_
    best_estimator_params = grid_search.best_params_
    
    def scorer(est, x, y):
        return np.sqrt(np.sum((est.predict(x) - y) ** 2) / len(y))
    
    scores = cross_val_score(best_estimator, X, Y, cv=10, scoring=scorer)
    
    return {
        'estimator': best_estimator, 
        'estimator_params': best_estimator_params,
        'scores': scores
    }
```

```python
def regressors_report(regressors, params, data):
    '''
       regressors - list of sklearn estimators implementing fit and predict methods
       params - list of sets of various parameters for GridSearch
       data - list of tuples (x, y) for fitting and predicting
    '''
    
    cros_val_results = dict()
    regressors_info = dict()
    
    for regressor, param, (X, Y) in zip(regressors, params, data):
        regressor_name = classifier_name(regressor)
        regr = regressor_prepaire(regressor, param, X, Y)
        
        regressors_info[regressor_name] = regr
        
        cros_val_results[regressor_name] = regr['scores']
        
    cros_val_results = pd.DataFrame(data=cros_val_results)
    
    return {
        'regressors': regressors_info,
        'report': cros_val_results
    }
```

### Comparing scores for model fitted by experts and automatically selected features 

```python
random_forest_grid_params = {
    'n_estimators': list(range(140, 150, 1)),
    'max_depth': list(range(11, 15, 1)), 
}
```

```python
trains = dict()

for column in report_features['report'].columns:
    column_features = list(report_features['report'][column].get_values())
    estimator_name = column.split('_')[0]
    
    estimator_data = train[column_features].get_values()
    estimator_target = train[['critical_temp']].get_values()
    
    trains[estimator_name] = (estimator_data, estimator_target)

scores_report_by_rfe = regressors_report([ElasticNetCV, LassoLarsCV, RandomForestRegressor], 
                                         [{}, {}, {}, random_forest_grid_params], 
                                         [trains['ElasticNetCV'], 
                                          trains['LassoLarsCV'], 
                                          trains['RandomForestRegressor']])
```

### RMSE for models

```python
for estimator_name, estimator_content in scores_report_experts['regressors'].items():
    estimator = estimator_content['estimator']
    train[estimator_name + '_prediction'] = estimator.predict(train_experts.get_values()[:, :-1])
    
    crit_temp = train['critical_temp']
    predicted_temp = train[estimator_name + '_prediction']
    
    RMSE = np.sqrt(np.sum((crit_temp - predicted_temp) ** 2) / len(crit_temp))
    
    print('RMSE ', estimator_name, ' is ', round(RMSE, 3))
```
