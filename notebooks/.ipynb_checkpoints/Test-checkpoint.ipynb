## Plot & analyze several training examples of each class

```python
import seaborn as sns
import numpy as np
sns.set()
from matplotlib import pyplot as plt
%matplotlib inline
```

#### Preparation

```python
tar_ind=dict()
for target in target_set:
    tar_ind[target]=train[train['Target']==target].index
```

```python
def rand_tar_index(target):
    '''Returns random index from the target class'''

    return np.random.choice(tar_ind[target])
```

```python
def f_plot(y, legend):
    '''Common plot drawer'''
    
    #plt.figure(figsize=(25,5))
    plt.plot(y, label=legend)
    plt.xticks(range(0,len(y),50));
    plt.xlim(0,len(y))
    plt.legend();
```

#### Random examples visualization

```python
n_plots=2 #number of plots, even only
plt.figure(figsize=(25,5))

for i in range(int(n_plots/2)):
    for target in target_set:
        f_plot(train.loc[rand_tar_index(target)], str(target))
```

```python
n_plots=4 #number of plots, even only

plt.figure(figsize=(25,5))
for i in range(int(n_plots/2)):
    for tar in target_set:
        row = rand_tar_index(tar)
        window = train.columns[1:]
        y = train.loc[row, window]
        label = str(train.loc[row, 'Target'])
        f_plot(y, label)
        
        #plt.plot(train.loc[rand_tar_index(tar)], label=str(tar))
```

# Feature hunt

#### Preparation

```python
def feat_dist(feat, timepoints, values=False, plot=True):
    '''Feature distribution for each class'''
    #feat = func(series)
    if values==False and plot==False:
        return print('Warning: Useless start of feat_dist()')
    
    feat_class_dict={}
    for tar in set(train.loc[:,'Target']):
        feat_class_list=list()
        for ind in tar_ind[tar]:
            feat_class_list.append(feat(train.loc[ind, timepoints]))
        feat_class_dict[tar]=feat_class_list
        
    if plot==True:
        for tar in feat_class_dict.keys():
            sns.distplot(feat_class_dict[tar], label=str(tar))
            plt.legend()
    
    if values==True:
        return(feat_class_dict)
```

### Let's do it

### Analysis of discrete Fourier Transform (FT)

```python
plt.figure(figsize=(25,5))
for tar in target_set:
    fourier = abs(np.fft.fft(ts_mean.loc[tar]))
    plt.title('Absolute values of FT')
    f_plot(fourier, str(tar))
    plt.xlim((0,60))
```

```python
n_plots=4 #number of plots, even only

plt.figure(figsize=(25,5))
for i in range(int(n_plots/2)):
    for tar in target_set:
        row=rand_tar_index(tar)
        window=train.columns[1:]
        fourier = abs(np.fft.fft(train.loc[row,window]))
        plt.title('Absolute values of FT')
        f_plot(fourier, str(tar))
        plt.xlim((0,60))
```

#### Hypothesis: there is difference between abs values of TF for timesteps from 40 to 60
#### Proposed feature: sum of abs values of FT for timesteps from 40 to 60

```python
def sum_abs_FT(series):
    '''Sum of abs values of FT for timesteps from 30 to 50'''
    
    fourier = abs(np.fft.fft(series))[40:60]
    return np.sum(fourier)
```

```python
feat_dist(sum_abs_FT, train.columns[1:])
plt.xlim(0,1000)
```

```python
def FT_rule(series):
    flag=sum_abs_FT(series)
    if flag>330 or flag<70:
        return -1
    else:
        return 1
```

### Analysis of the gradient

```python
plt.figure(figsize=(25,5))
for tar in target_set:
    f_plot(np.gradient(ts_mean.loc[tar]), str(tar))
```

```python
n_plots=2 #number of plots, even only
plt.figure(figsize=(25,5))

for i in range(int(n_plots/2)):
    for target in target_set:
        y=np.gradient(train.loc[rand_tar_index(target)])
        label=str(target)
        f_plot(y, label)
```

#### Hypothesis: plot of the gradient for class -1 is much more unstable. 
#### Proposed feature: number of extremums of the gradient

```python
def grad_ext_num(series):
    '''Counter of extremums of the series gradient'''
    
    counter=0
    d_grad=np.gradient(np.gradient(series))
    for i in range(1, len(d_grad)):
        if np.sign(d_grad[i])!=np.sign(d_grad[i-1]):
            counter+=1 
    return counter
```

```python
feat_dist(grad_ext_num, train.columns[1:])
```

```python
def grad_rule(series):
    flag=sum_abs_FT(series)
    if flag>78 or flag<51:
        return -1
    else:
        return 1
```

#### Checking other degrees of gradient

```python
def ngrad_ext_num(series):
    '''Counter of extremums of the series gradient in power of n'''
    
    counter=0
    d_grad=series
    for k in range(n):
        d_grad=np.gradient(d_grad)
    for i in range(1, len(d_grad)):
        if np.sign(d_grad[i])!=np.sign(d_grad[i-1]):
            counter+=1 
    return counter
```

```python
def ngrad_rule(series):
    n=4
    flag=ngrad_ext_num(series)
    if flag>97:
        return -1
    else:
        return 1
```

# Accuracy of the proposed rules

```python
from sklearn.metrics import accuracy_score
```

```python
for row in test.index:
    test.loc[row,'FT'] = FT_rule(test.loc[row,'Time-0':'Time-499'])
    test.loc[row,'4grad'] = ngrad_rule(test.loc[row, 'Time-0':'Time-499'])
```

```python
FT_accuracy = accuracy_score(test.Target, test.FT)
grad_accuracy = accuracy_score(test.Target, list(test.loc[:,'4grad']))
```

```python
print('\t     Accuracy\n',
      35*'_',
      '\n FT_rule: \t |{0}'.format(FT_accuracy),
      '\n grad_rule: \t |{0}'.format(grad_accuracy))
```
