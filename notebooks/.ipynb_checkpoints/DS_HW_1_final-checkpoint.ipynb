## Choosing the features

### Feature 1: number of extreme points of a times series. According to the picture above, it seems that objects from class "-1" period is larger than the same of object from class "1". Therefore, number of extreme points might be the feature.

```python
def count_extr(series):
    
    'Calculation of extreme points amount in a certain row'
    
    sign_changes = np.sign(np.gradient(series))
    count = 0

    for s_1, s_2 in zip(sign_changes[:-1], sign_changes[1:]):
        if s_1 * s_2 < 0:
            count += 1
            
    return count
```

```python
one_count = list()

for row_num in range(one.shape[0]):
    count = count_extr(one[row_num, 1:])
    one_count.append(count)
```

```python
minus_one_count = list()

for row_num in range(minus_one.shape[0]):
    count = count_extr(minus_one[row_num, 1:])
    minus_one_count.append(count)
```

```python
one_extr_number = sum(one_count) / len(one_count)
minus_one_extr_number = sum(minus_one_count) / len(minus_one_count)
```

```python
print('Average number of extreme points for objects of class "-1" : ', one_extr_number)
print('Average number of extreme points for objects of class "1" : ', minus_one_extr_number)
```

```python
def class_accuracy(data, mid):
    
    'Verification of accuracy of classification by \\
    calculation of extreme points number and comparison with dividing value '

    ans = list()

    for row_num in range(data.shape[0]):
        content = data[row_num, 1:]
        CLASS = data[row_num, 0]

        count = count_extr(content)
        cl = 0

        if count < mid:
            cl = 1
        else:
            cl = -1

        ans.append( int(CLASS == cl) )

    return sum(ans) / len(ans)
```

#### Â Selection of an optimal dividing value for a number of extreme points

```python
correct_ans_train = list()
correct_ans_test = list()

MID_MIN = 45
MID_MAX = 75

for mid in range(MID_MIN, MID_MAX):
    ans = class_accuracy(train_array, mid)
    correct_ans_train.append(ans)
    
    ans = class_accuracy(test_array, mid)
    correct_ans_test.append(ans)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(correct_ans_train, 'orange', label='accuracy_train(mid)')
plt.plot(correct_ans_test, 'g', label='accuracy_test(mid)')
plt.grid()
plt.legend()
plt.show()
```

### Feature 4. Expansion into Fourier series (FFT)

```python
plt.figure(figsize=(15,5))
for target in [-1, 1]:
    fourier = abs(np.fft.fft(train_mean.loc[target]))
    plt.title('Absolute values of FT')
    custom_plot(fourier, str(target))
    plt.xlim((0,100))
```

#### On the picture the Fourier series transformation for mean values is shown. Easy to see that all the peaks are situated between 20 and 60 and classes are visually different and can be divided. We can consider sum of coefficients in Fourier series transformation as a feature and pick out critical values for classification. 

```python
def sum_abs_fft(series):
    '''Sum of abs values of FT for timesteps'''
    
    left = 40
    right = 60
    
    fourier = abs(np.fft.fft(series))[left:right]
    return np.sum(fourier)
```

#### After a series of experiments an area for sum calculation was chosen: Time-40 - Time-60

```python
feature_dist(sum_abs_fft, train.columns[1:])
plt.xlim(0,1000)
```

#### Distribution of Fourier coefficients sum values for different classes. It is possible to select a tendency or a rule, that: "-1" - (0, 90) and (380, 1000)

## Creating the rules and calculation of accuracy with test

### As a result, only features 1 and 4 fit. It is necessary to make rules for them, define classes with the help of these rules and count the accuracy

```python
def ft_feature(series):
    feature = sum_abs_fft(series)
    if feature > 380 or feature < 90:
        return -1
    else:
        return 1
```

```python
def extr_rule(series, mid):
    feature = count_extr(series)
    if feature > mid:
        return -1
    else:
        return 1
```

```python
for row in test.index:
    test.loc[row,'fft'] = ft_feature(test.loc[row,'Time-0':'Time-499'])
    test.loc[row,'extremes'] = extr_rule(test.loc[row, 'Time-0':'Time-499'], 68)
```

```python
fft_feature = accuracy_score(test.Target, list(test.loc[:,'fft']))
extr_feature = accuracy_score(test.Target, list(test.loc[:,'extremes']))
```

```python
print('Accuracy extreme feature: {}'.format(extr_feature))
print('Accuracy fft feature: {}'.format(fft_feature))
```
