# Problem: prediction of biopsy results

Biopsy is 100% cancer detector. Wiki: "It is a medical test commonly performed by a surgeon, interventional radiologist, or an interventional cardiologist involving extraction of sample cells or tissues for examination to determine the presence or extent of a disease."  
It is important to make preliminary prediction if it nececcary to make this test because it expencive and extremely unpleasant for the patient.

### Data import

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
sns.set()
%matplotlib inline
```

If you have no internet access, read database from file with replacing url to path

```python
url='http://archive.ics.uci.edu/ml/machine-learning-databases/00383/risk_factors_cervical_cancer.csv'
data=pd.read_csv(url)
```

```python
data.head()
```

## Preprocessing 

We are able not to encode the 'yes'/'no' answers because they are represented as binary values.  
But after importing there are a huge number of string values in the table. It has happened because of '?' values (people prefered not to answer) take place instead of None.  
Fixing this problems with converting '?' to None and string values to float.

```python
def float_converter(x):
    if x=='?':
        return np.NaN
    else:
        return float(x)
    
data=data.applymap(float_converter)
```

```python
data.info()
```

None values mean that people prefered not to answer.   
Adding new columns if people answered the questions and fill nan with median values.   
Also, there are almost empty columns named 'STDs: Time since first diagnosis' and 'STDs: Time since last diagnosis' which I decided to drop.

```python
data.drop(['STDs: Time since first diagnosis','STDs: Time since last diagnosis'], axis=1, inplace=True)
```

```python
data['Sexual partners answer'] = data['Number of sexual partners'].map(lambda x: x==x).map({True:1, False:0})
data['First sex']=data['First sexual intercourse'].map(lambda x: x==x).map({True:1, False:0})
data['Pregnancues answer']=data['Num of pregnancies'].map(lambda x: x==x).map({True:1, False:0})
data['Smokes answer'] = data['Smokes'].map(lambda x: x==x).map({True:1, False:0})
data['Hormonal Contraceptives answer']=data['Hormonal Contraceptives'].map(lambda x: x==x).map({True:1, False:0})
data['IUD answer']=data['IUD'].map(lambda x: x==x).map({True:1, False:0})
data['STDs answer']=data['STDs'].map(lambda x: x==x).map({True:1, False:0})
```

```python
data=data.fillna(data.median())
```

```python
data.head()
```

```python
data.info()
```

Also, there is something wrong with values: 

```python
list(set(data['Smokes (packs/year)']))[:2]
```

Fixing this problem with rounding (converting float to int):

```python
data=data.applymap(lambda x: round(x))
```

```python
list(set(data['Smokes (packs/year)']))
```

### Making count plots 
to see the classes distribution 

```python
for col in data.columns:
    g = sns.catplot(x=col, hue='Biopsy', kind='count',
                data=data)
    g.fig.set_size_inches(20,5)
```

We can see that the classes are disbalanced. Features 'STDs:cervical condylomatosis' and 'STDs:AIDS' are totally useless. Drop them

```python
data.drop(['STDs:cervical condylomatosis', 'STDs:AIDS'], axis=1, inplace=True)
```

Next step is the correlation check

```python
plt.figure(figsize=(25,20))
sns.heatmap(data.corr(), cmap=plt.cm.Greens);
```

Check features with maximal correlation coefficient for Biopsy

```python
plt.figure(figsize=(15,13))
sns.heatmap(data.loc[:,['Dx:Cancer','Dx:HPV', 'STDs:HIV', 
                        'STDs:genital herpes','Hinselmann',
                        'Schiller','Citology','Biopsy']].corr(), annot=True, cmap=plt.cm.Greens);
```

### Biopsy result strongly depends on the results of the 'Hinselmann', 'Schiller' and 'Citology'. It is reasonable to use them for prediction model

```python
y=data['Biopsy']
X=data.loc[:,'Hinselmann':'Citology']
```

#### Make one split at the first step

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=420)
```

```python
f, axes = plt.subplots(1, 2, figsize=(15, 4), sharex=True)
sns.despine(left=True)

tar_train_count=y_train.value_counts()
tar_train_count.plot(kind='bar', title='Count (train target)', ax=axes[0]);

tar_test_count=y_test.value_counts()
tar_train_count.plot(kind='bar', title='Count (test target)', ax=axes[1]);
```

#### Try decision tree classifier (because of easy interpretation) with default parametres using one split

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier()
tree_clf.fit(X_train, y_train)
```

```python
y_pred = tree_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

print('\t     Quality:\n',
      35*'_',
      '\n f1_score: \t |{0}'.format(f1_score(y_test, y_pred)),
      '\n accuracy_score: |{0}'.format(accuracy_score(y_test, y_pred)))
```

#### Visualisation the quality. It is clear that accuracy_score does't suit us (but f1_score does)

```python
from sklearn.metrics import confusion_matrix

conf_mat = confusion_matrix(y_true=y_test, y_pred=y_pred)
print('Confusion matrix:\n', conf_mat)

labels = ['Class 0', 'Class 1']
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
cax = ax.matshow(conf_mat, cmap=plt.cm.Greens)
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('Expected')
plt.show()
```

### Using cross_validation: StratifiedKFold instead of Kfold because of classes disbalance.

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
```

```python
clf = DecisionTreeClassifier()
Kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=420)
```

### Selection the classifier parameters using GridSearch   
(class_weight is extremely important because of classes disbalance)

```python
from sklearn.model_selection import GridSearchCV
```

```python
tree_param = {
    'criterion': ['gini', 'entropy'],
    'max_depth': list(range(1, 15, 1)),
    'splitter': ['random', 'best'],
    'max_features': list(range(1,3,1)),
    'min_samples_leaf': list(range(1,5,1)),
    'presort': [True, False],
    'class_weight': [{0:0.01,1:5},
                     {0:0.01,1:1},
                     {0:0.1,1:1},
                     {0:1,1:1}],
    'min_samples_split': list(range(2,40,2))
}
```

```python
tree_grid_search = GridSearchCV(DecisionTreeClassifier(), tree_param, n_jobs=4, cv=Kfold, scoring='f1')
tree_grid_search.fit(X_train,y_train)
tree_best_params = tree_grid_search.best_estimator_
y_pred_tree = tree_grid_search.predict(X_test)
print('{0}\nf1_score={1}'.format(tree_best_params, f1_score(y_test, y_pred_tree)))
```

```python
conf_mat = confusion_matrix(y_true=y_test, y_pred=y_pred_tree)
print('Confusion matrix:\n', conf_mat)

labels = ['Class 0', 'Class 1']
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
cax = ax.matshow(conf_mat, cmap=plt.cm.Greens)
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('Expected')
plt.show()
```
